```python
from IPython.display import HTML
import uuid

def mermaid(code: str):
    div_id = f"mermaid_{uuid.uuid4().hex}"
    return HTML(f"""
    <div id="{div_id}" class="mermaid">
    {code}
    </div>
    <script>
    if (typeof mermaid !== 'undefined') {{
        mermaid.init(undefined, "#{div_id}");
    }}
    </script>
    """)
```

```python
mermaid("""
graph TD
    A[Start] --> B{Decision}
    B -->|Yes| C[Proceed]
    B -->|No| D[Stop]
""")
```
