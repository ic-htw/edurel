# EdurelDB

```python
con = ddbu.file_con(DB_DIR + "db-edurel", read_only=False)
```

```python
sql = """
create table Prompt (
  pid integer not null,
  tag text not null,
  lang text not null,
  db text not null,
  prompt text not null,
  constraint pk_prompt primary key (pid),
);
"""
con.execute(sql)
```
