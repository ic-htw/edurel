```python
con = ddbu.mem_con(f"{DB_DIR}/db-company_en")
schema = ddbu.schema(con)
```

```python
url = 'https://wigpu01l.f4.htw-berlin.de:4000/api/generate'
# Benutzername und Passwort für die Basic Authentication
username = 'iclassen'
password = 'ollama1rx2'

headers = {
    'Content-Type': 'application/json'
}
payload = {
    "model": "a-kore/Arctic-Text2SQL-R1-7B",
    "prompt": f"Explain the following database schema:\n{schema}\n",
    "stream": False
}

# POST-Anfrage an die API senden (mit Authentifizierung)
response = requests.post(url, json=payload, headers=headers, auth=(username, password), verify=False)  # verify=False ignoriert SSL-Zertifikat

# Antwort überprüfen
if response.status_code == 200:
    print("API-Aufruf erfolgreich:")
    try:
        # Lade JSON-Antwort und formatiere sie mit dem UTF-8 Zeichensatz
        response_data = response.json()
        formatted_response = json.dumps(response_data, indent=4, ensure_ascii=False).encode('utf-8')
        # print(formatted_response.decode())

        # Response formatiert ausgeben
        print(response_data['response'])  # Direkter Zugriff auf den Text der Antwort
    except json.JSONDecodeError as e:
        print(f"Fehler beim JSON-Decoding der Antwort: {e}")
else:
    print("Fehler beim API-Aufruf:")
    print(response.status_code, response.text)
```
